```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Section A
np.random.seed(42)
data = np.random.randint(0, 100, size=(1000, 6))
df = pd.DataFrame(data, columns=['C/C++', 'Python', 'MATLAB', 'R', 'Assembly', 'Java'])
print("--- Section A: Data ---")
print(df.head())

# Section B
def get_experience_score(student_vector):
    return np.sum(student_vector)
    first_student = df.iloc[0].values
    score = get_experience_score(first_student)
    print(f"\n--- Section B: Score ---")
    print(f"Total experience: {score}")

    # Section C
    def classify_student(student_vector, w, b):
        z = np.dot(student_vector, w) + b
        return "Has taken" if z >= 0 else "Hasn't taken"
        weights = np.array([0.2, 0.3, 0.1, 0.1, 0.1, 0.2])
        bias = -150
        result = classify_student(first_student, weights, bias)
        print(f"\n--- Section C: Result ---")
        print(f"Classification: {result}")

            # Section D
        X_2d = df[['C/C++', 'Python']].values
        w_2d = np.array([0.2, 0.3])
        b_2d = 30
        plt.figure(figsize=(8, 6))
        plt.scatter(X_2d[:, 0], X_2d[:, 1], alpha=0.3, label='Students')
        x_range = np.linspace(0, 100, 100)
        y_boundary = (-(w_2d[0] * x_range) - b_2d) / w_2d[1]
        plt.plot(x_range, y_boundary, color='red', label='Decision Boundary', linewidth=3)
        plt.legend()
        plt.title('Section D: 2D Visualization')
        plt.show()
```
